```python
import pandas as pd
```

```python
df = pd.read_csv('bls_occupation_allocc.csv')
```

```python
df.columns = df.iloc[0]
df = df[1:]
df = df.reset_index(drop=True)
```

```python
df
```

```python
major = df[df['Occupation type'] == 'Summary']

major = major[major['2023 National Employment Matrix code'].str[3] == '0']
major.loc[0, 'Occupation type'] = 'Total'
major
```

```python
df_allocc = df[df['Occupation type'] == 'Line item']
df_allocc['Major Occupation Group Code'] = df_allocc['2023 National Employment Matrix code'].str[:2]

major_occupations = {'11': 'Management', '13': 'Business and Financial Operations', '15': 'Computer and Mathematics', '17': 'Architecture and Engineering', '19': 'Life, Physical, and Social Science', '21': 'Community and Social Services', '23':'Legal', '25':
 'Educational Instruction and Library', '27': 'Arts, Design, Entertainment, Sports, and Media', '29': 'Healthcare Practitioners and Technical', '31': 'Healthcare Support', '33': 'Protective Service', '35': 'Food Preparation and Serving', '37': 'Building and Grounds Cleaning and Maintenance', '39': 'Personal Care and Service', '41':'Sales',
  '43':'Office and Administrative Support', '45':'Farming, Fishing, and Forestry', 
  '47':'Construction and Extraction', '49':'Installation, Maintenance, and Repair', 
  '51':'Production', '53':'Transportation and Material Moving'}

df_allocc['Major Occupation Group'] = df_allocc['Major Occupation Group Code'].map(major_occupations)
df_allocc
```

```python
cleaned_data = pd.concat([major, df_allocc])
cleaned_data = cleaned_data.drop(columns=['Major Occupation Group Code', 'Related Occupational Outlook Handbook (OOH) content'])
cleaned_data = cleaned_data.reset_index(drop=True)
cleaned_data
```

```python
cleaned_data['Employment, 2023'] = cleaned_data['Employment, 2023'].str.replace(',', '')
cleaned_data['Employment, 2033'] = cleaned_data['Employment, 2033'].str.replace(',', '')
cleaned_data['Employment change, numeric, 2023–33'] = cleaned_data['Employment change, numeric, 2023–33'].str.replace(',', '')
cleaned_data['Occupational openings, 2023–33 annual average'] = cleaned_data['Occupational openings, 2023–33 annual average'].str.replace(',', '')
cleaned_data['Median annual wage, dollars, 2023[1]'] = cleaned_data['Median annual wage, dollars, 2023[1]'].str.replace(',', '')

cleaned_data['Percent self employed, 2023'] = cleaned_data['Percent self employed, 2023'].str.replace('—', '0.0')
cleaned_data['Median annual wage, dollars, 2023[1]'] = cleaned_data['Median annual wage, dollars, 2023[1]'].str.replace('—', '0.0')
cleaned_data['Median annual wage, dollars, 2023[1]'] = cleaned_data['Median annual wage, dollars, 2023[1]'].str.replace('>=$239200', '239200.0')


cleaned_data
```

```python
convert_dict = {'Employment, 2023': float, 'Employment, 2033': float, 'Employment distribution, percent, 2023':float,
       'Employment distribution, percent, 2033':float,
       'Employment change, numeric, 2023–33':float,
       'Employment change, percent, 2023–33':float, 'Percent self employed, 2023':float,
       'Occupational openings, 2023–33 annual average':float,
       'Median annual wage, dollars, 2023[1]':float}

# Convert columns using the dictionary
cleaned_data= cleaned_data.astype(convert_dict)
```

```python
cleaned_data['Percent self employed, 2023'] = cleaned_data['Percent self employed, 2023'].replace(0, None)
cleaned_data['Median annual wage, dollars, 2023[1]'] = cleaned_data['Median annual wage, dollars, 2023[1]'].replace(0, None)
```
