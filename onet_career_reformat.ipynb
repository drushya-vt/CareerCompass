```python
import pandas as pd
```

```python
onet = pd.read_json("career_data_0302.json")
```

```python
onet = onet.T
onet = onet.reset_index()
```

```python
onet = onet.rename(columns={"index": "occupational_code"})
```

```python
onet
```

```python
onet_overview = onet[["occupational_code", "job_description", "alternate_titles", "tools_used", "technical_skills", "core_tasks"]]
onet_overview.to_csv("onet_overview.csv", index=False)
```

```python
skills = onet[["occupational_code", "skills"]]
skills.head()
```

```python
skills_expanded = skills.explode("skills")
skills_expanded = skills_expanded.reset_index()
skills_expanded
```

```python
skill = []
importance = []

for i in range(len(skills_expanded["skills"])):
    skill_data = skills_expanded["skills"][i]
    if isinstance(skill_data, dict):  # Check if the element is a dictionary
        skill.append(skill_data.get("Soft Skills", None))  # Use .get() to avoid KeyError
        importance.append(skill_data.get("Importance", None))
    else:
        skill.append(None)  # Append None if the element is not a dictionary
        importance.append(None)
skills_df = pd.DataFrame(
    {"occupational_code": skills_expanded["occupational_code"], "skill": skill, "importance": importance})
skills_df
```

```python
skills_df.to_csv("onet_skills.csv", index=False)
```

```python
# Split the occupational_code by '.' and filter rows where the last part is '00'
onet_filtered = onet[onet['occupational_code'].str.split('.').str[1] == '00']

onet_filtered["occupational_code"] = onet_filtered["occupational_code"].str.split('.').str[0]

# Reset the index of the filtered DataFrame
onet_filtered = onet_filtered.reset_index(drop=True)

onet_filtered 
```

```python
onet_filtered.to_csv("onet_overview.csv", index=False)
```

```python
skill = []
importance = []

skills = onet_filtered[["occupational_code", "skills"]]
skills_expanded = skills.explode("skills")
skills_expanded = skills_expanded.reset_index()

for i in range(len(skills_expanded["skills"])):
    skill_data = skills_expanded["skills"][i]
    if isinstance(skill_data, dict):  # Check if the element is a dictionary
        skill.append(skill_data.get("Soft Skills", None))  # Use .get() to avoid KeyError
        importance.append(skill_data.get("Importance", None))
    else:
        skill.append(None)  # Append None if the element is not a dictionary
        importance.append(None)
skills_df = pd.DataFrame(
    {"occupational_code": skills_expanded["occupational_code"], "skill": skill, "importance": importance})
skills_df
```

```python
education = []
importance = []

educations = onet_filtered[["occupational_code", "education"]]
education_expanded = educations.explode("education")
education_expanded = education_expanded.reset_index()

for i in range(len(education_expanded["education"])):
    education_data = education_expanded["education"][i]
    if isinstance(education_data, dict):  # Check if the element is a dictionary
        education.append(education_data.get("Education Level", None))  # Use .get() to avoid KeyError
        importance.append(education_data.get("Importance Level", None))
    else:
        education.append(None)  # Append None if the element is not a dictionary
        importance.append(None)
education_df = pd.DataFrame(
    {"occupational_code": education_expanded["occupational_code"], "education": education, "importance": importance})
education_df
```

```python
skills_df.to_csv("onet_skills.csv", index=False)
education_df.to_csv("onet_education.csv", index=False)
```
